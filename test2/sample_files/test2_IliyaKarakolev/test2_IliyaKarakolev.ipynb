```python
import glob
import pandas

# Merge the two data sets

def read_csv(pathnames):
    with open(pathnames, 'r') as csvfile:
        return pd.read_csv(csvfile, delimiter=csvDelimiter, encoding='utf-8')
pathnames = glob.glob('C:/Users/IliyaKarakolev/Desktop/test2/sample_files/test2_IliyaKarakolev/sample_*.csv')

csvDelimiter = ";"
decimalDelimiter_dest='.'

csv_files_with_pathnames = [(read_csv(pathname), pathname) for pathname in pathnames]


for csv, pathname in csv_files_with_pathnames:
    csv.columns = pd.MultiIndex.from_tuples(list(zip((pathname, ''), csv.columns)))
    
dest_file = 'C:/Users/IliyaKarakolev/Desktop/test2/sample_files/test2_IliyaKarakolev/merged_data_sets.csv'

merge = pd.concat([csv for csv, pathname in csv_files_with_pathnames], axis=1)
merge.to_csv(dest_file, sep=',',  decimal=decimalDelimiter_dest)
# I have to find out how to properly separate the first columns at each data set.
```
